```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Load the example flights dataset from seaborn
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

# Initialize the figure
fig, ax = plt.subplots(figsize=(10, 8))

# Plotting function
def animate(i):
    ax.clear()
    sns.heatmap(flights.iloc[:, :i+1], cmap="coolwarm", cbar=True, ax=ax)
    ax.set_title(f'Year: {flights.columns[i]}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Month')

# Animate the heatmap
ani = FuncAnimation(fig, animate, frames=len(flights.columns), interval=500, repeat=True)

# Save the animation as a gif
ani.save('heatmap_animation.gif', writer=PillowWriter(fps=2))

plt.show()
```
